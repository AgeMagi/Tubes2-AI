```python
## import pandas

import pandas as pd
import numpy as np
```

```python
# Path menuju file dataset training
file_path = './tubes2_HeartDisease_train.csv'

# Header kolom
headers = ['age', 'sex', 'CPT', 'RBP','SC', 'FBS','rECG', 'MHR', 'EIA','STDI','PE_ST', 'n_MV','thal', 'diagnosis']

# Membaca dataset training
train_dataset = pd.read_csv(file_path, header=None, engine='python', skiprows=1, na_values='?', names=headers)
train_dataset.head(15)
```

# Preprocessing Dataset Train

## Get data to fill Nan dataset train

```python
mean_age = train_dataset['age'].mean()
mode_sex = train_dataset['sex'].mode()
mode_cpt = train_dataset['CPT'].mode()
mode_rbp = train_dataset['RBP'].mode()
mean_sc = train_dataset['SC'].mean()
mode_fbs = train_dataset['FBS'].mode()
mode_recg = train_dataset['rECG'].mode()
mean_mhr = train_dataset['MHR'].mean()
mode_eia = train_dataset['EIA'].mode()
mean_stdi = train_dataset['STDI'].mean()
mode_pest = train_dataset['PE_ST'].mode()
mean_nmv = train_dataset['n_MV'].mean()
mode_thal = train_dataset['thal'].mode()

fill_dataset = {
    'age': mean_age,
    'sex': mode_sex,
    'CPT': mode_cpt,
    'RBP': mode_rbp,
    'SC': mean_sc,
    'FBS': mode_fbs,
    'rECG': mode_recg,
    'MHR': mean_mhr,
    'EIA': mode_eia,
    'STDI': mean_stdi,
    'PE_ST': mode_pest,
    'n_MV': mean_nmv,
    'thal': mode_thal,
}

fill_dataset = pd.DataFrame(data = fill_dataset).iloc[0]
display(fill_dataset)
```

## Fill dataset train

```python
train_dataset_filled = train_dataset.fillna(fill_dataset)
train_dataset_filled.head(15)
```
